```python
import pennylane as qml
```

```python
data = qml.data.load("qspin", sysname="Ising", periodicity="full", lattice="rectangular", layout="4x4")
data[0].list_attributes()
data[0].spin_system
```

```python
for item in data:
    print(item.list_attributes())
    print(item.parameters)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for i in range(len(data)):
    ax.plot(data[i].parameters['h'],data[i].order_params)

ax.set_xlabel('External magnetic field in X direction, h')
ax.set_ylabel('Magnetization in the Z direction')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

plot_data = pd.DataFrame([data[0].parameters['h'],data[0].order_params],columns=['h','z'])
sns.set_theme(style="whitegrid")

sns.displot(
    data=plot_data,
    x="h", hue="cut",
    kind="kde", height=6,
    multiple="fill", clip=(0, None),
    palette="ch:rot=-.25,hue=1,light=.75",
)
```
